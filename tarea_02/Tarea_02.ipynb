```python
# Problema 01:

# Se solicita imprimir los números del -100 al 200, por lo que se indica que 'a' son los números en el rango solicitado [en este caso del -100 al 201 al el programa considerar el 0] y se imprime, dando como resultado todos los números dentro del rango.

# Código:

for a in range (-100,201):
  print (a)
```

```python
# Problema 02:

# Se pone una condición para imprimir 'b' [un rango de números] con el 'if [si]', en este caso que sean números divisibles por 6 y que den resto 0 ['%' ayuda a conseguir el resultado al ser el símbolo utilizado para realizar divisiones que entreguen el resto].

# Código:

for b in range (-100, 201):
  if b % 6 == 0:
    print(b)
```

```python
# Problema 03:

# El código pide dos números al usuario [int, para formato números, con input para que el usuario escriba] [ya que se quiere hacer un código para números en general], el programa los suma y muestra un mensaje dependiendo si el resultado es menor a 200, entre 200 y 249, o mayor a 250 [condiciones].

# Código:

numero1 = int(input("Primer número: "))
numero2 = int(input("Segundo número: "))

suma = numero1 + numero2

if suma < 200:
    print("menor a 200")
elif 200 <= suma < 250:
    print("mayor a 200")
else:
    print("mayor a 250")
```

```python
# Problema 04:

# Se define [def] un código que evalúa la edad y el gusto musical del usuario. Pide los datos [input], convierte la respuesta sobre música a un valor booleano [tipo de dato lógico que solo puede tener dos posibles valores: verdadero o falso. En este caso se indica como verdadero el 'sí'[se pone 'lower' para transformar todas las respuestas en minúsculas y evitar errores de tipeo], si se responde otra cosa es 'falso' o 'no'], y luego imprime un mensaje personalizado según si el usuario es mayor de edad y si le gusta la música urbana [condiciones para el mensaje].

# Código:

def evaluar_usuario(edad, gusta_urbana):
    if edad >= 18 and gusta_urbana:
        print("¡Qué buen gusto!")
    elif edad >= 18 and not gusta_urbana:
        print("¡Debes tener otros gustos interesantes!")
    elif edad < 18 and gusta_urbana:
        print("Joven y con buenos gustos")
    else:
        print("Joven y con diversos gustos")

edad = int(input("¿Cuántos años tienes?: "))
musica = input("¿Te gusta la música urbana? (sí/no): ")

gusta_urbana = musica.lower() == "sí"

evaluar_usuario(edad, gusta_urbana)
```
